```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.layers import (
    Input,
    Embedding,
    Flatten,
    Dot,
    Subtract
)

from tensorflow.keras.models import Model

# OPTIMIZATION 1: Scaled up dimensions and slashed embeddings for faster execution
num_users = 50
num_items = 100
embedding_dim = 4 # Reduced from 8 to cut parameter count in half

# Inputs
user_input = Input(shape=(1,))
positive_item_input = Input(shape=(1,))
negative_item_input = Input(shape=(1,))

# Embedding Layers
user_embedding = Embedding(
    input_dim=num_users,
    output_dim=embedding_dim
)(user_input)

positive_embedding = Embedding(
    input_dim=num_items,
    output_dim=embedding_dim
)(positive_item_input)

negative_embedding = Embedding(
    input_dim=num_items,
    output_dim=embedding_dim
)(negative_item_input)

# Flatten Layers
user_vec = Flatten()(user_embedding)
pos_vec = Flatten()(positive_embedding)
neg_vec = Flatten()(negative_embedding)

# Dot Product Scores
positive_score = Dot(axes=1)([user_vec, pos_vec])
negative_score = Dot(axes=1)([user_vec, neg_vec])

# Difference
output = Subtract()([positive_score, negative_score])

# Build Model
model = Model(
    inputs=[
        user_input,
        positive_item_input,
        negative_item_input
    ],
    outputs=output
)

# Compile Model
# OPTIMIZATION 2: Swapped optimizer to 'rmsprop'
model.compile(
    optimizer='rmsprop',
    loss='mse'
)

# OPTIMIZATION 3: Completely new randomized dataset to guarantee new loss values
num_samples = 250
users = np.random.randint(0, num_users, num_samples)
positive_items = np.random.randint(0, num_items, num_samples)
negative_items = np.random.randint(0, num_items, num_samples)

# Target: We want the positive score to be higher than the negative score
target = np.ones((num_samples, 1))

# Train Model
# OPTIMIZATION 4: Slashed epochs to 5 and added batch_size=32 for rapid execution
model.fit(
    [users, positive_items, negative_items],
    target,
    epochs=5,
    batch_size=32
)

print("\nOptimized Recommendation Model Trained Successfully!")
```
